```python
import pandas as pd
import numpy as np
```

```python
math_course = pd.read_csv(
    filepath_or_buffer="student-mat.csv"
)
```

```python
math_course
```

```python
mapper = {
    'GP': "Gabriel Pereira",
    'MS': "Mousinho da Silveira"
}

math_course["school"] = math_course["school"].replace(mapper)
```

```python
math_course
```

```python
(
    math_course
    .groupby(by="school")
    .agg(number_of_students=("school", "count"))
)
```

```python
mapper = {
    'M': "Male",
    "F": "Female"
}

math_course["sex"] = math_course["sex"].replace(mapper)
```

```python
(
    math_course
    .groupby(by="school")
    .agg(
        number_of_students=("school", "count"),
        number_of_boys=("sex", lambda x: (x == "Male").sum()),
        number_of_girls=("sex", lambda x: (x == "Female").sum())

    )
)
```

#### `sex` and `address` distribution

```python
mapper = {
    'U': "Urban",
    "R": "Rural"
}

math_course["address"] = math_course["address"].replace(mapper)
```

```python
(
    math_course
    .groupby(by="sex")
    .agg(
        number_of_students=("sex", "count"),
        from_rural=("address", lambda address: (address == "Rural").sum()),
        from_urban=("address", lambda address: (address == "Urban").sum())
    )
)
```
